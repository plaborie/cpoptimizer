```python
import json
with open("./RCPSP_utils/"+instance+".red.json") as data_file:
    data = json.load(data_file)
n = data["ntasks"]
m = data["nresources"]
C = data["capacities"]
D = data["durations"]
R = data["requirements"]
S = data["successors"]

GFILE="../RCPSP/RCPSP_utils/"+instance+".graph"
with open(GFILE, 'w') as outfile: 
    outfile.write(str(n)+' '+str(len(S))+'\n')
    for s in S:
        outfile.write(str(s[0])+' '+str(s[1])+'\n')
        
import sys,os
exec='/Users/laborie/MyOffice/Dev-GIT/COS-HEAD-BIS/cpoptimizer/lib/x86-64_osx/static_pic/transitive_reduction '
print('Transitive reduction: RUNNING ...')
os.system(exec + GFILE + ' ' + str(nbIterations) + ' ' + str(seed))
print('Transitive reduction: DONE.')

SRED = []
with open(GFILE+'.red', 'r') as infile:
    l1 = infile.readline().split()
    ns = int(l1[1])
    for i in range(ns):
        pr = infile.readline().split()
        p = int(pr[0])
        s = int(pr[1])
        SRED.append([p,s])

nbRemoved = len(S)-len(SRED)
ratio =  100.0 * nbRemoved / len(S)
print('Transitive reduction removed ' + str(round(ratio,2)) + '% of arcs')

data = { "ntasks":n, 
         "nresources":m, 
         "capacities":C, 
         "durations":D, 
         "requirements":R, 
         "successors": SRED
       }

import json
with open("../RCPSP/RCPSP_utils/"+instance+".red.json", 'w') as outfile: 
    json.dump(data, outfile)
```

If the problem is small, we can display the set of precedence constraints between tasks as a directed graph.

```python
import networkx as nx
from nxpd import draw
G = nx.DiGraph(rankdir='LR',dpi=50)
G.add_nodes_from(range(n))
G.add_edges_from(SRED)
draw(G,show='ipynb')
```

```python
# Import Constraint Programming modelization functions
from docplex.cp.model import *
model = CpoModel()

start = [integer_var(min=0, max=10000000, name='S'+str(i)) for i in range(n)]
end   = [integer_var(min=0, max=10000000, name='S'+str(i)) for i in range(n)]

# Decision variables: tasks
for i in range(n):
    model.add(end[i]-start[i]==D[i])

# Constraints: precedence between tasks 
for s in S:
    model.add(end[s[0]] <= start[s[1]])
```

## Solving the problem with CP Optimizer automatic search

The model can be solved by calling:

```python
# Solve the model
model.export_model("./RCPSP_utils/"+instance+".cpo")
sol = model.solve(LogPeriod=1000000,TimeLimit=60,Workers=1,trace_log=True)
#sol = model.solve(LogPeriod=1000000000,FailureDirectedSearchEmphasis=4,TimeLimit=900,trace_log=True)
```
